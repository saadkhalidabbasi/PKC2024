 ### EDA Stands for Exploratory Data Analysis

Exploratory Data Analysis (EDA) is a method used to `analyze` and `summarize` datasets.In Python, you can use the Pandas Library to perform EDA on DataSet

```python
# Step-1 Importig Libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step-2 Import DataSet
df=sns.load_dataset('titanic')
```

```python
df.drop(columns='deck',inplace=True)
```

```python
df.shape
```

## Visualzing our Data and Exploring through Graphs

```python
# Visualzing our Data 
sns.countplot(data=df,x='survived')
```

## For Bar Plot , we should take categorical value on one axis and numeric variable on other axis 

```python
sns.barplot(data=df,x='survived',y='fare',hue='who')
```

```python
sns.countplot(data=df,x='pclass')
```

```python
# Data Distribution
sns.histplot(data=df,x='age',bins=4)

# data distribution is at a specific time how many units of data are present 
```

### We can only make hist plot of numeric variables only 

```python
df['age'].hist(bins=30,color='darkred',alpha=0.75)
```

## For Making Box Plot ,we should take numeric on one axis and Categorical value on other axis


## We CannOT Make box plots of two numeric variables

```python
# Box Plot 
sns.boxplot(data=df,y='age',x='survived', hue='who')
```

## We see and analyize the outliers of our data through Box Plot 

```python
# If we want to place our both numeric vairables on both axis , we wil use the Scatter Plot 

# Scatter Plot 
sns.scatterplot(data=df,x='age',y='fare',hue='pclass')

plt.ylim(0,300)

# we can set limit of our data by using matplotlib library 
```

```python
sns.pairplot(df,hue='survived')
```

## Stacked Bar Chart/plot

```python
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

### SWARM Plot

```python
sns.swarmplot(data=df,x='survived',y='fare' ,hue='who')
```

```python
sns.swarmplot(data=df,x='survived',y='age',hue='who')
```

### Point Plot

```python
sns.catplot(x='pclass',y='fare',hue='embarked',kind='point',data=df)
```

```python
# Relation of two numeric variable 

sns.lmplot(data=df,y='fare',x='age',hue='pclass')
plt.ylim(0,300)
```

```python
# How to add a column and make a age group 
# Grouping

df['age_groups']=pd.cut(df['age'],[0,10,18,30,50,80],labels=['Kids','Child','Youth','Middle_Age','Old'])

# we have added a new group or columns based on age variable .
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby('age_groups')['survived'].mean().plot(kind='bar')
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
# we can make a pair plot of our own choose variable like below 

sns.pairplot(df, hue='survived',vars=['age','fare','pclass'])



```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# importing data set 

kashti=sns.load_dataset('titanic')
kashti
```

```python
kashti.info()
```

```python
# Checking Nul values 
kashti.isnull().sum()
```

##  Bar Plot

```python
kashti.survived.value_counts().plot(kind='bar')
```

```python
sns.countplot(x='survived',hue='who', data=kashti)
```

```python
sns.countplot(x='survived',hue='sex',data=kashti)
```

### Scattter Plot 

```python
# Scatter Plot 

plt.scatter(data=kashti,x='age', y='fare')
```

```python
# For Making the scatter plot , both the axis values should be numeric 

sns.scatterplot(data=kashti,x='age',y='fare',hue='pclass')
```

```python
sns.barplot(data=kashti,x='sex',y='fare',hue='pclass')
```

```python
# For A bigger version of Graph, We Make A pair Graph 

sns.pairplot(kashti)
```

## Hist Plot 

```python
kashti.hist()
```

```python
kashti.hist(bins=10,figsize=(9,7),grid=False)
```

## Box Plot 

```python
sns.boxplot(data=kashti,x='sex',y='fare',hue='class')
```

```python
sns.boxenplot(data=kashti,x='sex',y='fare',hue='embark_town')
```

```python
# Changing the titles of X and Y axis .

sns.scatterplot(data=kashti, x='fare',y='age',hue='sex')
```

```python
# Making A Grid Plot 
g=sns.FacetGrid(kashti,hue='survived',col='pclass',margin_titles=True,palette={1:'seagreen',0:'black'})

g=g.map(plt.scatter,"age","fare",edgecolor='w').add_legend()
```

```python
g=sns.FacetGrid(kashti,hue='survived',col='sex',margin_titles=True,palette={1:'seagreen',0:'black'})

g=g.map(plt.scatter,"age","fare",edgecolor='w').add_legend()
```

```python
sns.stripplot(data=kashti,x='sex',y='fare',hue='class')
```

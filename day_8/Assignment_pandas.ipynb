### How to Rename Columns

```python
df=pd.DataFrame({'a':[1,2,3,],'b':[3,4,5],'c':[6,7,8],'d':[10,11,12]})
df.columns
df.head()
```

```python
df.rename(columns={'a':'z','b':'y','c':'x','d':'u'}, inplace=True)
df
```

```python
prac=pd.DataFrame(np.random.rand(5,10),columns=list('ABCDEFGHIJ'))
prac.head()
```

```python
prac.rename(columns={'A':'X'})
```

```python
saad=pd.DataFrame([[1,2],[2,4]],columns=list('ac'))
saad
```

```python
saad.add_prefix('a')
```

```python
saad.add_suffix('z')
```

## Reverse Row order 

```python
import pandas as pd 
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# Reversing Order of Rows

df.loc[::-1]
```

```python
# Reversing order of columns
df.loc[:,::-1]
```

```python
# Resetting the previous changes of rows and columns change 

df.loc[::-1].reset_index(drop=True)
```

```python
df.loc[:,::-1].reset_index(drop=True)
```

```python
# Selecting a specific data type of columns 

df.select_dtypes(include='float64')
```

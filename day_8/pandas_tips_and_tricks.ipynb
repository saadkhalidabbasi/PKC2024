# Split DataFrames into two subsets 

```python
import pandas as pd 
import seaborn as sns 
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
# finding lenght of data set 

len(df)

# it wil return the no of rows 
```

```python
df.shape
```

```python
kashti_1=df.sample(frac=0.5,random_state=1)
kashti_1.shape
kashti_1
# It splitted the original data set having 244 rows into half . i.e 122 rows 
```

```python
# Similarly 
kashti_2=df.drop(kashti_1.index)
kashti_2.shape
```

```python
# len is used to find the length of dataset 
df=len(kashti_1)+len(kashti_2)
df
```

```python
kashti_1.shape
```

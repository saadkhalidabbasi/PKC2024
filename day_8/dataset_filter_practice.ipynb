# **Split Data Frames into two subsets** 

```python
# 1- Load data set to be splitted 
split_data_set=sns.load_dataset('tips')
split_data_set
```

```python
# 2- Calculating the no of entries 

len(split_data_set)
```

```python
# Now we will split our data set into two subset each having 122 entries or 50% entries 

first_split_section=split_data_set.sample(frac=0.5,random_state=1)
first_split_section

# we got 122 random entries due to random state 
```

```python
# Now again calculating the 2nd half or 2nd 50% of the dataset 

second_data_set=split_data_set.drop(first_split_section.index)
second_data_set
```

```python
# Now checking the total entries of each newly dataset 
first_split_section.shape
```

```python
second_data_set.shape
```

## Join Two data Set 

```python
# We can join two spllitted dataset back or two random dataset togather .

# Syntax :

# newdataset =pd.concat([firstdataset,seconddataset],ignore_index=False)

combined_data_Set=pd.concat([first_split_section,second_data_set],ignore_index=False)


combined_data_Set.shape

# We have joined back the two different dataset togather
```

## **Filtering the DataSet**

```python
# We can filter the dataset based on different features 

# For Filtering , we will take a new dataset from seaborn library

import seaborn as sns 
import pandas as pd
```

```python
filter_data_set=sns.load_dataset('titanic')
filter_data_set.head()
```

```python
# We will check the unique values in a specific column
# Syntax for checking the unique values 

# datasetname.column-name.unique()

filter_data_set.embark_town.unique()
```

```python
# Filtering based on a specific values if any 

# data-set-name[(data-set-name.column-name==='specific value')]


filter_data_set[(filter_data_set.embark_town=='Southampton')]

# It is showing only that result that have southampton 
```

```python
# Similarly 
# We can also apply more filter based on column values 
# We can also apply multiple conditions 
# For example 

filter_data_set[(filter_data_set.embark_town=='Southampton') &
                (filter_data_set.sex=='female')
                
                ]

# It is showing only females that are board from southampton
```

## Grouping 


```python
flower.groupby(['species']).describe()
```

```python
# Taking transpose and grouping  
flower.groupby(['species']).describe().T
```

```python
flower=flower.drop(columns='Unnamed: 0')
```

```python
flower
```

```python
# Grouping the species column 
# Syntax of Grouping 

# datasetname.groupby(['column_name']).describe()

flower.groupby(['species']).describe()
```

```python
# Taking Transpose of the above grouping 

flower.groupby(['species']).describe().T
```

```python
# Count The Number of species by sepal_length 
flower.groupby(['species','sepal_length']).count()
```

```python
flower.groupby(['species','sepal_length','petal_length'])['species'].count()
```

```python
flower.count()
```

```python
# Making a CountPlot 
sns.countplot(x='species',data=flower)
```

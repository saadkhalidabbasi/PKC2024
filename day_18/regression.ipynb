```python
# model evaluation  metrices 
# r2 
# mean squred error 
# mean absolute error 
# root mean squred error


x_train, x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
# Again calling model 
model=LinearRegression()

model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
# making a plot to compare the y_Test and prediction values to check the accuracy 
sns.scatterplot(x=y_test,y=prediction)
plt.xlabel('Actual Value')
plt.ylabel('predict value')

plt.show()
```

```python
print('Shape of y_test:', y_test.shape)
print('Shape of prediction:', prediction.shape)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
MAE=mean_absolute_error(y_true=y_test,y_pred=prediction)
MSE=mean_squared_error(y_true=y_test,y_pred=prediction)
RMSE=mean_squared_error(y_true=y_test,y_pred=prediction,squared=False)
R2=r2_score(y_true=y_test,y_pred=prediction)


print("MAE=",MAE)
print("MSE=",MSE)
print("RMSE=",RMSE)
print("R2=",R2)
```

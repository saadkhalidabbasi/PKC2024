```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
df.tail()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(data=df,x='sepal_width')
```

```python
df_new=df[['sepal_length','sepal_width','petal_length']]
```

```python
df_new
```

```python
x=df_new
```

```python
y=df[['petal_width']]
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[6.2,3.4,5.4]])
```

```python
model.predict([[4.4,3.2,1.3]])
```

```python

x_train, x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
x_train.shape
```

```python
pred = np.squeeze(pred)

y_test = np.squeeze(y_test)
```

```python
sns.scatterplot(x=pred,y=y_test)
```

```python
model.score(x_test,y_test)
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df=sns.load_dataset('iris')
df.head()
df.tail()
```

```python
df.isnull().sum()
```

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
x.head()
```

```python
y=df['species']
```

```python
model=DecisionTreeClassifier()
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
model.fit(x_train,y_train)
```

```python
model.predict([[5.8,2.3,1.2,4.3]])
```

```python
per=model.predict(x_test)
```

```python
cmm=confusion_matrix(y_test,per)
```

```python
sns.heatmap(cmm,annot=True)
plt.ylabel('Actual Values')
plt.xlabel('Predicted Values ')
plt.show()
```

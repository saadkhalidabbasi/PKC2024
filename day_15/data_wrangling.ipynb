```python
# Import Libraries 
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

We can see that our column "sepal_Width has some outlier " we can definetlty remove them 

```python
# Checking the Normaliation of our dataset 
sns.histplot(data=df,x='sepal_length',kde=1)
```

```python
sns.histplot(data=df,x='sepal_width',kde=1)
```

```python
sns.histplot(data=df,x='petal_length',kde=1)
```

```python
sns.histplot(data=df,x='petal_width',kde=1)
```

```python
df.shape
```

```python
Q1=df['petal_length'].quantile(0.25)
Q3=df['petal_length'].quantile(0.75)
IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR


df=df[(df['petal_length']>lower_limit)&(df['petal_length']<upper_limit)]
```

```python
df.shape
```

```python
Q1=df['sepal_width'].quantile(0.25)
Q3=df['sepal_width'].quantile(0.75)
IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR


df=df[(df['sepal_width']>lower_limit)&(df['sepal_width']<upper_limit)]
```

```python
df.shape
```

```python
Q1=df['sepal_length'].quantile(0.25)
Q3=df['sepal_length'].quantile(0.75)
IQR=Q3-Q1

lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR


df=df[(df['sepal_length']>lower_limit)&(df['sepal_length']<upper_limit)]
```

```python
df.shape
```
